## Background and Prior Work

Several studies have established a robust link between critical reception and box office performance, demonstrating a positive correlation between favorable reviews and revenue. For example, the study "Predicting Box Office Success: Do Critical Reviews Really Matter?" highlights this correlation, particularly for films with wide releases. It also underscores the pivotal role of marketing campaigns in enhancing a movie's success and its box office potential (Kennedy). Furthermore, the study "Do Critical Reviews Affect Box Office Revenues Through Community Engagement and User Reviews?" explores how the valence of critical reviews impacts sales both directly and through user engagement metrics like community participation and the volume of user reviews. The results indicate that while critical reviews do have a positive relationship with box office revenue, this connection is partially mediated by factors like User reviews volume and community interaction(Chiu, Du, Wang, Sun).

More specifically, the study "How Critical Are Critical Reviews? The Box Office Effects of Film Critics, Star Power, and Budget" reveals that films such as "Batman Returns," despite high production costs, achieved significant first-week revenue. This analysis indicates that negative reviews have a more detrimental impact on revenue than the boost provided by positive reviews. Additionally, it shows that while star power and large budgets can mitigate the effects of negative reviews, they are less effective when reviews are predominantly positive, emphasizing the significant roles of critical reviews, marketing, and star presence in determining a film's financial success (Basuroy, Chatterjee, Ravid).

Given these insights, there is an opportunity to develop a range of machine learning models, including linear regression, multiple regression, and stepwise regression models, utilizing features such as but not limited to marketing budget, star power, competitiveness, critical reception, and pre-release buzz. We will calculate metrics like the number of competitive releases and average star power to enhance our model. Our objective is to compare how different models perform based on these selected features and identify the most effective model for prediction. We aim to achieve a Mean Squared Prediction Error within 10% and strive for a robust R-squared value. This model will evaluate the influence of key factors such as critical reception and marketing expenditures on revenue, aiming to use these features to predict future box office revenue accurately. This approach will provide comprehensive insights into how movie companies can adjust their strategies to maximize revenue.

## Data Cleaning and Wrangling

### Data Cleaning

After loading the data, we proceed by cleaning the data as follows:

- Transform columns into usable formats
  - Stringified numbers are converted to numeric values (float or int)
  - Stringified lists of JSON objects are parsed into lists of strings
    - Some special cases for this are for languages and countries where we keep only the standardized codes for each rather than their given name. This is to ensure that the subsequent analyses are readable and consistent
  - Stringified booleans are converted to boolean values
  - Any failed conversions for a given column are set to 0-like values for the relevant datatype (e.g. np.NaN for numeric values or empty array ([]) for lists)
- Remove rows that have missing (e.g. null), invalid (e.g. np.NaN), or 0-like values in columns of interest
- Keep movies only adhering to the following criteria:
  - Non-adult: the reason being that the adult category includes only about 9 movies and may contain inappropriate data for comparison with mainstream films
  - Released: we only want to examine movies that have actually been released to the public. Unreleased movies do not have box office revenue data, which is a key variable in our analysis
  - Movies released from 1990 onwards: we want to focus on modern movies
  - Original language is English: this is too avoid any potential differences in box office revenue due to language barriers
- Keep only columns of interest, and reorder as follows such that revenue is on the far right: ['id', 'title', 'overview', 'release_date', 'budget', 'genres', 'popularity', 'production_companies', 'production_countries', 'runtime', 'spoken_languages', 'vote_average', 'vote_count', 'revenue']

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the paths to the datasets
THE_MOVIES_DATASET_PATH =  'data/the-movies-dataset-rounak-banik/'
CLEANED_DATA_FILE_PATH = 'data/cleaned_movies_metadata.pkl'

# Set configurations
pd.set_option('display.max_columns', 50)
```

```python
# Read csv file into pandas
_credits = pd.read_csv(THE_MOVIES_DATASET_PATH + 'credits.csv')

keywords = pd.read_csv(THE_MOVIES_DATASET_PATH + 'keywords.csv')
links = pd.read_csv(THE_MOVIES_DATASET_PATH + 'links.csv')

links_small = pd.read_csv(THE_MOVIES_DATASET_PATH + 'links_small.csv')
movies_metadata = pd.read_csv(THE_MOVIES_DATASET_PATH + 'movies_metadata.csv')

ratings = pd.read_csv(THE_MOVIES_DATASET_PATH + 'ratings.csv')
ratings_small = pd.read_csv(THE_MOVIES_DATASET_PATH + 'ratings_small.csv')
```

```python
"""
Define some utility functions
"""

# Attempt to convert release_date to datetime, put NaN if str invalid
def to_datetime_safe(date_str):
    # If already converted to datetime, return as is
    if type(date_str) == pd.Timestamp:
        return date_str
        
    try:
        return pd.to_datetime(date_str)
    except:
        return np.nan


# Convert non-numeric values to numeric, put NaN if invalid
def to_numeric_safe(value):
    try:
        return pd.to_numeric(value)
    except:
        return np.nan
    
# Convert stringified  list of JSON objects of form:
#  [ {...,"name": "<NAME>",...}, {...,"name": "<NAME>",...}, ... ]
# to list of strings, put empty list if invalid
def to_list_safe(json_str):
    # If already converted to list, return as is
    if type(json_str) == list:
        return json_str

    try:
        return [i['name'] for i in eval(json_str)]
    except:
        return []   
    
def lang_to_list_safe(json_str):
    # If already converted to list, return as is
    if type(json_str) == list:
        return json_str

    # Retain only the iso_639_1 code
    try:
        return [i['iso_639_1'] for i in eval(json_str)]
    except:
        return []


"""
Convert stringified  list of JSON objects of form:
[ {...,"iso_3166_1": "<CODE>",...}, {...,"iso_3166_1": "<CODE>",...}, ... ]
to list of strings, put empty list if invalid

Note:
- ISO 3166-1 codes are standardized codes for countries
- this is a special case for country conversion, to improve readability since 
some countries have long names
"""
def country_to_list_safe(json_str):
    # If already converted to list, return as is
    if type(json_str) == list:
        return json_str

    # Retain only the iso_3166_1 code,
    try:
        return [i['iso_3166_1'] for i in eval(json_str)]
    except:
        return []
```

```python
"""
Transform columns to usable format
"""

mm_df = movies_metadata.copy()

# adult
mm_df['adult'] = mm_df['adult'].replace(
    {'True': True, 'False': False}).astype(bool)

# belongs_to_collection

# budget
mm_df['budget'] = mm_df['budget'].apply(to_numeric_safe)

# genres
mm_df['genres'] = mm_df['genres'].apply(to_list_safe)

# homepage
# id
mm_df['id'] = mm_df['id'].apply(to_numeric_safe)
mm_df['id'] = mm_df['id'].astype('Int64')  # Convert to integer

# imdb_id
# original_language
# original_title
# overview

# popularity
mm_df['popularity'] = mm_df['popularity'].apply(to_numeric_safe)

# poster_path

# production_companies
mm_df['production_companies'] = mm_df['production_companies'].apply(
    to_list_safe)
# production_countries
mm_df['production_countries'] = mm_df['production_countries'].apply(
    country_to_list_safe)

# release_date
mm_df['release_date'] = mm_df['release_date'].apply(to_datetime_safe)

# revenue
mm_df['revenue'] = mm_df['revenue'].apply(to_numeric_safe)

# runtime
mm_df['runtime'] = mm_df['runtime'].apply(to_numeric_safe)

# spoken_languages
mm_df['spoken_languages'] = mm_df['spoken_languages'].apply(lang_to_list_safe)

# status

# tagline

# title

# video

# vote_average
mm_df['vote_average'] = mm_df['vote_average'].apply(to_numeric_safe)
# vote_count
mm_df['vote_count'] = mm_df['vote_count'].apply(to_numeric_safe)

mm_df.shape

# trasforms strings to list for cast and crew in credits
_credits['cast'] = _credits['cast'].apply(to_list_safe)
_credits['crew'] = _credits['crew'].apply(to_list_safe)

# trasforms keywords strings to list for keywords
keywords['keywords'] = keywords['keywords'].apply(to_list_safe)
```

```python
"""
Drop missing, invalid, and 0-like values
"""

# Drop NaN values for relevant columns
mm_df = mm_df.dropna(subset=['release_date', 'status', 'original_language', 'adult', 'budget', 'popularity', 'revenue', 'runtime', 'vote_average', 'vote_count'])

# Drop rows that do not have a valid boolean value for boolean cols
mm_df = mm_df[mm_df['adult'].isin([True, False])]

# Drop rows that have a 0 value for numeric cols
mm_df = mm_df[mm_df['budget'] > 0]
mm_df = mm_df[mm_df['popularity'] > 0]
mm_df = mm_df[mm_df['revenue'] > 0]
mm_df = mm_df[mm_df['runtime'] > 0]
mm_df = mm_df[mm_df['vote_average'] > 0]
mm_df = mm_df[mm_df['vote_count'] > 0]

# Drop rows that have an empty array for array cols
mm_df = mm_df[mm_df['genres'].apply(lambda x: len(x) > 0)]
mm_df = mm_df[mm_df['production_companies'].apply(lambda x: len(x) > 0)]
mm_df = mm_df[mm_df['production_countries'].apply(lambda x: len(x) > 0)]
mm_df = mm_df[mm_df['spoken_languages'].apply(lambda x: len(x) > 0)]

# Note, we are only dropping data for columns that will be used for analysis /
# further processing
```

```python
# Keep only movies that meet aforementioned criteria
mm_df = mm_df[
    (mm_df['status'] == 'Released') & # Released movies 
    (mm_df['original_language'] == 'en') & # Made originally in English
    (~mm_df['adult']) & # Not adult movies
    (mm_df['release_date'].dt.year >= 1990) # Released after 1990
]
```

```python
# Keep only relevant columns, and reorder them into the following sections:
# movie metadata (id, title, etc.), production factors (budget, genres, etc.), 
# revenue
mm_df = mm_df[['id', 'title', 'overview', 'release_date', 'budget', 'genres', 'popularity', 'production_companies', 'production_countries', 'runtime', 'spoken_languages', 'vote_average', 'vote_count', 'revenue']]

print(f"Cleaned dataset shape: {mm_df.shape}")
mm_df.head(5)
```

### Data Wrangling

```python
# Merge in the cleaned keywords and credits to movies_metadata
mm_df = pd.merge(mm_df, keywords, on='id')
mm_df = pd.merge(mm_df, _credits, on='id')
```

```python
# Make a new dataframe with the average rating of each movie and the movieId
average_ratings_df = ratings.groupby('movieId').mean()
average_ratings_df['number_of_ratings'] = ratings.groupby('movieId').count()['rating']

# Add colunm tmdbId to average_ratings_df
# Use links_df to add the tmdbId column to average_ratings_df this is matched 
# based ont he movieId
links.dropna(subset=['tmdbId'], inplace=True)
links['tmdbId'] = links['tmdbId'].astype(np.int64)
average_ratings_df = pd.merge(average_ratings_df, links, on='movieId')

# Remove timestamp, userId, which are ununsed columns from ratings_df
average_ratings_df.drop(columns=['timestamp', 'userId'], inplace=True)

# Rename movieId to id
average_ratings_df.rename(columns={'tmdbId':'id'}, inplace=True)

# Merge the number of ratings and average ratings into the the mm_df
average_ratings_df.reset_index(inplace=True)
mm_df = pd.merge(mm_df, average_ratings_df, on='id')
```

```python
# Drop the columns that are not needed
mm_df.drop(columns=['movieId', 'imdbId', 'index'], inplace=True)
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]


# Remove outliers for each relevant column
columns_to_check = ['budget', 'revenue', 'runtime', 'vote_average',
                    'vote_count', 'rating', 'number_of_ratings', 'popularity']
for column in columns_to_check:
    mm_df_outliers = remove_outliers(mm_df, column)
```

#### Most Common Values 

To determine the most common values for each categorical variable, we counted the frequency of each unique value and sorted them in descending order. This analysis provides insights into the distribution of these categorical variables and highlights the most prevalent categories.

```python
def plt_most_common_vals(df, column, rank=10, plot=True):
    # Obtain value counts for each unique value in the column
    value_cnts = (df[column]
                  .explode()
                  .value_counts()
                  .sort_values(ascending=False)
                  )

    # Prepare data for plotting
    top_10_values = value_cnts.head(rank)
    vals = top_10_values.values
    labels = top_10_values.index

    if not plot:
        return top_10_values

    # Plot results
    sns.barplot(x=vals, y=labels, hue=labels, palette='viridis')
    plt.title(f"Top {rank} Most Common Values in \'{column}\'")
    plt.xlabel("Count")
    plt.ylabel(f"{column}")
    plt.show()

    return top_10_values
```

#### Highest Grossing Values

To determine which values are associated with the highest revenue, we calculated the total revenue for each unique value in the categorical variables. This analysis provides insights into which categories are associated with higher revenue.

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plt_highest_grossing_vals(df, column, rank=10, plot=True):
    # Explode the column to handle lists of values (e.g., cast, crew, keywords)
    exploded_df = df[[column, 'revenue']].explode(column)

    # Filter out the values that appear in more than 10 movies
    value_counts = exploded_df[column].value_counts()
    filtered_values = value_counts[value_counts >= 0].index
    filtered_df = exploded_df[exploded_df[column].isin(filtered_values)]

    # Group by the column and calculate total revenue
    top_vals_by_revenue = (filtered_df
                           .groupby(column)
                           .sum()
                           .sort_values(by='revenue', ascending=False)
                           .head(rank))

    # Prepare data for plotting
    top_vals_by_revenue['revenue'] /= 1e6  # convert to millions for readability
    vals = top_vals_by_revenue['revenue'].values
    labels = top_vals_by_revenue.index

    if not plot:
        return top_vals_by_revenue

    # Plot results
    sns.barplot(x=vals, y=labels, hue=labels, palette='viridis')
    plt.title(f"Top {rank} {column} Values by Total Revenue (in millions)")
    plt.xlabel("Total Revenue (in millions)")
    plt.ylabel(f"{column}")
    plt.show()

    return top_vals_by_revenue
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]


# Remove outliers for each relevant column
columns_to_check = ['budget', 'revenue', 'runtime', 'vote_average',
                    'vote_count', 'rating', 'number_of_ratings', 'popularity']
for column in columns_to_check:
    mm_df_outliers = remove_outliers(mm_df, column)
```

#### Transforming Categorical Variables to Numeric Vectors

We then transform our categorical variables into numeric vectors using **one-hot encoding** proceeded by **dimensionality reduction**, specifically **Principal Components Analysis (PCA)**. This process allows us to convert categorical variables into a numeric vector that can be used in our machine learning models. Furthermore, the use of PCA as an additional step beyond one-hot encoding ensures that we are not employing exceedingly sparse feature vectors which can have a deleterious effect on model performance.

```python
# mm_df = mm_df_outliers
# # First make a new dataframe that is a one hot encoding of the genres, production_countries, spoken_languages, keywords, cast, and crew columns
# # This will allow us to use these columns in the model

# # One hot encoding of genres
# genres_df = mm_df['genres'].str.join('|').str.get_dummies()

# # One hot encoding of production_countries
# production_countries_df = mm_df['production_countries'].str.join(
#     '|').str.get_dummies()

# # One hot encoding of spoken_languages
# spoken_languages_df = mm_df['spoken_languages'].str.join('|').str.get_dummies()

# # One hot encoding of keywords
# keywords_df = mm_df['keywords'].str.join('|').str.get_dummies()

# # One hot encoding of cast
# cast_df = mm_df['cast'].str.join('|').str.get_dummies()

# # One hot encoding of crews
# crew_df = mm_df['crew'].str.join('|').str.get_dummies()

# # Concatenate all the one hot encoded dataframes

# one_hot_df = pd.concat([genres_df, production_countries_df,
#                        spoken_languages_df, keywords_df, cast_df, crew_df], axis=1)

# one_hot_df.head()
```

```python

def get_exploded_oh_enc(df, idx_col, var_col):
    df = df[[idx_col, var_col]]

    # Convert genres to one-hot encoding
    df = df.explode(var_col)
    df = df.pivot_table(index=idx_col, columns=var_col, aggfunc='size', fill_value=0)

    return df

# Join all categorical variables into a single dataframe
merged_oh_df = mm_df[['id']]
for var in CATEGORICAL_VARS:
    exploded_oh_var = get_exploded_oh_enc(mm_df, 'id', var)
    # Print shape
    print(f"Shape of {var} one-hot encoded dataframe: {exploded_oh_var.shape}")
    merged_oh_df = pd.merge(exploded_oh_var, merged_oh_df, left_index=True, right_index=True, how='inner')

merged_oh_df.head()
```

```python
# Perform PCA on the one hot encoded data
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

NUM_PCA_COMPONENTS = 3

# Standardize the data
scaler = StandardScaler()
data_std = scaler.fit_transform(merged_oh_df)

# Apply PCA
pca = PCA(n_components=NUM_PCA_COMPONENTS)
data_pca = pca.fit_transform(data_std)

print(f"Original shape: {merged_oh_df.shape}")
print(f"Transformed shape: {data_pca.shape}")

# Add the PCA data to the mm_df 
pca_df = pd.DataFrame(data_pca, columns=[f'pca_{i}' for i in range(NUM_PCA_COMPONENTS)])
# normalize the PCA columns
#pca_df = scaler.fit_transform(pca_df)

mm_df = pd.concat([mm_df, pca_df], axis=1)
mm_df.head(5)
```

```python
# Append pca columns to the features list
FEATURES += [f'pca_{i}' for i in range(NUM_PCA_COMPONENTS)]
```

### Model Exploration

We explored a variety of models for predicting revenue. In particular, we focused on the following models:

#### Baseline Model: Mean Revenue Estimation

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import cross_val_score
import statsmodels.api as sm

X = mm_df[FEATURES]
y = mm_df[TARGET]

# Initialize the dummy regressor baseline model (predicts the mean revenue)
mean_rgsr = DummyRegressor(strategy="mean")
scores = cross_val_score(mean_rgsr, X, y,
                         scoring='neg_mean_absolute_error', cv=10)
mean_mae = -scores.mean()
print(f'Mean Regressor average MAE across 10-folds: {mean_mae}')
```

#### Model I: Standard Linear Regression

```python
# Define constants for model evaluation
TRAIN_SZ = 0.8 # Note, this is only for residual analysis
K_FOLDS = 10
```

```python
# Instantiate model with default params and score it
from sklearn.model_selection import cross_val_score
linear_rgsr = LinearRegression()

# Cross-validation score
scores = cross_val_score(
    linear_rgsr, X, y, scoring='neg_mean_absolute_error', cv=K_FOLDS)
linear_rgsr_mae = -scores.mean()
print(f'Average MAE across 10-folds: {linear_rgsr_mae}')
```

```python
# Compute residiuals on model using train-test split, note this is
# distinct from cross-validation which we used to score the model
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=(1-TRAIN_SZ), random_state=42)

linear_rgsr.fit(X_train, y_train)

y_pred = linear_rgsr.predict(X_test)

residuals = y_test - y_pred

plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals)

plt.title('Residual Plot')

plt.xlabel('Predicted Values')

plt.ylabel('Residuals')

plt.axhline(y=0, color='r', linestyle='--')

plt.show()
```

The plot show no clear pattern and a constant spread across the range of predicted values, which indicates that the variance of the residuals is homoscedasticity. And residuals are symmetrically distribute around the zero line too, which suggest no significant bias. 

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, validation_scores = learning_curve(
    estimator=linear_rgsr,
    X=X,
    y=y,
    train_sizes=np.linspace(0.1, 1.0, 10),
    cv=K_FOLDS,
    scoring='neg_mean_absolute_error')

train_scores_mean = -train_scores.mean(axis=1)
validation_scores_mean = -validation_scores.mean(axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
         label="Training error")
plt.plot(train_sizes, validation_scores_mean, 'o-', color="g",
         label="Validation error")
plt.title('Learning curve')
plt.xlabel('Training set size')
plt.ylabel('MAE')
plt.legend(loc="best")
plt.show()
```

Initially, as more data is used for training shown in the red line, the training error typically decreases because the model has more information to learn from. On the other hand, the green line that represent the validation error usually starts high because the model is trained on a small sample size of of data and thus generalizes poorly, the MAE decreases as more data is used for improving generalization and  stabilizes, indicating the point where adding more training data doesn't significantly improve model performance on unseen data.

```python
from sklearn.linear_model import Ridge

ridge_rgsr = Ridge(alpha=1.0, solver='sag', random_state=42)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

scores = cross_val_score(ridge_rgsr, X_scaled, y,
                         scoring='neg_mean_absolute_error', cv=K_FOLDS)

ridge_rgsr_mae = -scores.mean()
print(f'Ridge Regressor average MAE across 10-folds: {ridge_rgsr_mae}')
```

```python
ridge_rgsr.fit(X_train, y_train)

y_pred = ridge_rgsr.predict(X_test)

residuals = y_test - y_pred

plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, color='royalblue')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residual Plot for Ridge Regression')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```

The residuals' patterns do not flcutuate a lot as predicted values increase, indicating consistent error variance across predicted values representing homoscedasticity. 

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    ridge_rgsr, X_scaled, y,
    train_sizes=np.linspace(0.1, 1.0, 10),
    cv=K_FOLDS, scoring='neg_mean_absolute_error')

train_scores_mean = -np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = -np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, 'o-',
         color="r", label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-',
         color="g", label="Cross-validation score")

plt.title('Learning Curve for Ridge Regression')
plt.xlabel('Training set size')
plt.ylabel('Mean Absolute Error')
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

This red line represents the MAE on the training data, which the error decreases sharply as the training set size increases initially and then begins to stable. This suggests that the model benefits significantly from more data at the start, but each additional data point contributes less to improving the model's performance once a certain amount of data is reached. On the other hand, the green line indicates the validation score shows how well the model generalizes to unseen data. It starts relatively high, indicating poor generalization when trained with small amounts of data. As more data is used, the validation MAE drops significantly, indicating better generalization. This line stabilizes and slightly increases towards the larger training sizes, which might indicate that the model starts to face limitations in learning more from the added data in the validation folds. 

Gap between two lines: Initially, there's a significant gap between the training and validation scores, suggesting overfitting when the model is trained with smaller data sets. As more data is used, the gap narrows significantly, indicating that the model is generalizing better.
